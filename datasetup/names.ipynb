```python
import random

random.seed(1234)
adjectives = [
    "Velvet", "Crimson", "Golden", "Silver", "Twilight", "Mystic", "Eclipse", "Sunset", 
    "Moonlit", "Radiant", "Serene", "Whispering", "Dancing", "Enchanted", "Ancient", "Majestic",
    "Royal", "Classic", "Vintage", "Elegant", "Bold", "Lush", "Rustic", "Vibrant",
    "Gentle", "Silent", "Harmonious", "Lively", "Playful", "Joyful", "Peaceful", "Soothing",
    "Exotic", "Wild", "Mystical", "Secret", "Hidden", "Lost", "Forgotten", "Legendary",
    "Dreamy", "Sensual", "Romantic", "Passionate", "Fiery", "Tempestuous", "Tranquil", "Melodic",
    "Autumn", "Winter", "Spring", "Summer", "Morning", "Noon", "Evening", "Midnight",
    "Northern", "Southern", "Eastern", "Western", "Celestial", "Galactic", "Cosmic", "Infinite",
    "Azure", "Emerald", "Ruby", "Sapphire", "Diamond", "Pearl", "Opal", "Jade",
    "Floral", "Fruity", "Spicy", "Earthy", "Woody", "Nutty", "Herbal", "Mineral"
]

nouns = [
    "Harbor", "Meadow", "Grove", "Orchard", "Vineyard", "Forest", "Valley", "Mountain",
    "River", "Sea", "Ocean", "Lake", "Spring", "Falls", "Creek", "Brook",
    "Castle", "Palace", "Fortress", "Chateau", "Estate", "Manor", "Villa", "Lodge",
    "Garden", "Park", "Oasis", "Sanctuary", "Haven", "Retreat", "Reserve", "Domain",
    "Journey", "Voyage", "Expedition", "Quest", "Adventure", "Odyssey", "Saga", "Tale",
    "Dream", "Vision", "Mirage", "Phantom", "Shadow", "Echo", "Whisper", "Murmur",
    "Flame", "Fire", "Blaze", "Ember", "Spark", "Light", "Star", "Sun",
    "Moon", "Comet", "Planet", "Galaxy", "Universe", "Nebula", "Meteor", "Asteroid",
    "Blossom", "Bloom", "Petal", "Thorn", "Leaf", "Branch", "Twig", "Root",
    "Spice", "Pepper", "Cinnamon", "Ginger", "Saffron", "Vanilla", "Mint", "Basil"
]

# Function to generate wine names
def generate_wine_names(adjectives, nouns, total_names):
    wine_names = set()
    while len(wine_names) < total_names:
        adj = random.choice(adjectives)
        noun = random.choice(nouns)
        wine_names.add(f"{adj} {noun}")
    return wine_names
```

```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")
df = df[df['quality'].isin([4, 5, 6, 7])]
unique_wine_names = generate_wine_names(adjectives, nouns, df.shape[0])
df.index = list(unique_wine_names)
df.to_csv("wine.csv")
```

```python
import plotly.express as px
from plotly.offline import plot
import plotly.graph_objs as go
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde
wine_data = pd.read_csv("wine.csv", index_col= 0)
# feature_name = "fixed acidity"
feature_name = "pH"
wine_quality = 5
fig = px.histogram(wine_data, x=feature_name, color="quality", marginal="rug", histnorm="percent", barmode ="overlay")

for trace in fig.data:
    trace.update(opacity=1)

for trace in fig.data:
    if int(trace.name) == wine_quality:
        vcolor = trace.marker.color
        break

# Add KDE lines for each quality
qualities = wine_data['quality'].unique()
for quality in qualities:
    quality_data = wine_data[wine_data['quality'] == quality][feature_name]
    kde = gaussian_kde(quality_data)
    x_range = quality_data.max() - quality_data.min()
    kde_line = go.Scatter(x=np.linspace(quality_data.min(), quality_data.max(), 200),
                        y=2*x_range * kde(np.linspace(quality_data.min(), quality_data.max(), 200)),
                        mode='lines',
                        line=dict(width=2),
                        showlegend=False)
    for trace in fig.data:
        if trace.name == quality:
            kde_line.line.color = trace.marker.color
            break

    fig.add_trace(kde_line)

plot_div = plot(fig)
```
